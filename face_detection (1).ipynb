```python
import numpy as np
import cv2
```

```python
from google.colab.patches import cv2_imshow
img=cv2.imread('friends.jpg')
cv2_imshow(img)
```

```python
#load cascade classifier
haar=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#step1-convert img to grayscale
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
#step 2- apply grayscale img to cascade
face_rect=haar.detectMultiScale(img_gray)
print(face_rect) # 5 rows=5 faces detected
```

```python
#step3-draw bounding box
image=img.copy()
#The computer already knows where the faces are (from some facefinding trick)
#Each face has a box around it with:
#x, y = where the box starts (top-left corner).
#w, h = how wide and tall the box is.
for x,y,w,h in face_rect:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),thickness=2)
cv2_imshow(image)
#0,255,0= tuple for green color
```

```python
import cv2
from IPython.display import display, Image, clear_output
import time
from google.colab.patches import cv2_imshow

# Read the video
cap = cv2.VideoCapture('video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Display frame using cv2_imshow
    cv2_imshow(frame)

    # Add delay (40ms = ~25fps) and clear previous frame
    time.sleep(0.04)
    clear_output(wait=True)

    # Exit if 'q' is pressed (doesn't always work in Colab)
    # if cv2.waitKey(1) & 0xFF == ord('q'):
    #     break

cap.release()
```
