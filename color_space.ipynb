# color space

*   color conversion
*   gray scale


```python
#gray=r+g+b/3
#using pillow,matplotlib,opencv
import matplotlib.image as matimg
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img_cv=cv2.imread('flying_bird.jpg')#bgr
```

```python
#numpy slicing
b=img_cv[:,:,0]
g=img_cv[:,:,1]
r=img_cv[:,:,2]
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
#using opencv channels
b,g,r=cv2.split(img_cv)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
#coverting color
img_gray=cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
plt.imshow(img_gray)
```

```python
img_color=cv2.cvtColor(img_gray,cv2.COLOR_GRAY2BGR)
plt.imshow(img_color)
```
